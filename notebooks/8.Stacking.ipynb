# Stacking

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report
```

```python
dataset = pd.read_csv("processed_reddit_data.csv")

cleaned_dataset = dataset.dropna()

X_cleaned = cleaned_dataset["clean_comment"]
y_cleaned = cleaned_dataset["category"]

X_train_cleaned, X_test_cleaned, y_train_cleaned, y_test_cleaned = train_test_split(
    X_cleaned, y_cleaned, test_size=0.2, random_state=42, stratify=y_cleaned
)

tfidf_cleaned = TfidfVectorizer(max_features=5000)
X_train_tfidf_cleaned = tfidf_cleaned.fit_transform(X_train_cleaned)

# basic learners
lightgbm_model = LGBMClassifier(
    objective="multiclass",
    num_class=3,
    random_state=42,
    metric="multi_logloss",
    is_unbalance=True,
    class_weight="balanced",
    reg_alpha=0.1,
    reg_lambda=0.1,
    learning_rate=0.1,
    n_estimators=100,
    max_depth=7,
)
logistic_model = LogisticRegression(max_iter=200, class_weight="balanced")

knn_model = KNeighborsClassifier(n_neighbors=5)

# stacking classifier
stacking_clf = StackingClassifier(
    estimators=[
        ("lgbm", lightgbm_model),
        ("logistic", logistic_model),
        ("knn", knn_model),
    ],
    final_estimator=LogisticRegression(class_weight="balanced"),
    cv=5,
)

stacking_clf.fit(X_train_tfidf_cleaned, y_train_cleaned)
y_pred_cleaned = stacking_clf.predict(tfidf_cleaned.transform(X_test_cleaned))
print(classification_report(y_test_cleaned, y_pred_cleaned))
```
